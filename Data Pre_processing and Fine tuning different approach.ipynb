```python
# Data Preprocessing
from bs4 import BeautifulSoup
import unidecode
import nltk 
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
import string
import re
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
stopwordsList = stopwords.words("english")
```

```python
def preprocessText(text):
   Html_free_text =  BeautifulSoup(text,"html.parser").text # extracting text from HTML if any inside HTml
   Accent_removed_text = unidecode.unidecode(Html_free_text) # accent text removal
   Special_char_removed_text = re.sub(r'[^a-zA-Z0-9\s]', '', Accent_removed_text)   # special charcters removal
   tokens = nltk.word_tokenize(Special_char_removed_text) 
   lemetized_text = ''
   for token in tokens:
      text = lemmatizer.lemmatize(token)  # doing lemmatization
      if not text in stopwordsList: # stop words removal
         if(text != ''):  # whitespace removal
            lemetized_text = lemetized_text +" "+text
   lower_case_text = lemetized_text.lower() # Converting into lowercase to normalize the text
   stripped_text = lower_case_text.strip() 
   translator = str.maketrans('', '', string.punctuation) # punctutaion removal step 1 making translator
   punctuation_removed_text = stripped_text.translate(translator) # punctuation removal using translate
   return punctuation_removed_text   # preprocessed text
```

```python
text_col = df["Text"]
pre_processed_text = text_col.apply(preprocessText)
```

```python
df["Text"] = pre_processed_text
```

```python
dataset = df.loc[:,["Text" , "Summary"]]
```

```python
# GPT-2 Model Intialization
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
model_name = 'gpt2'
output_dir = 'FineTunedModel'
overwrite_output_dir = False
per_device_train_batch_size = 8
num_train_epochs = 50.0
save_steps = 50000
```

```python
# Split the dataset 75:25
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(dataset["Text"],dataset["Summary"])
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# Data custom class creation
from transformers import TextDataset, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments
```

```python
def load_data_collator(tokenizer, mlm = False):
    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, 
        mlm=mlm,
    )
    return data_collator
```

```python
train_dataset = X_train + " TL;DR " + y_train
data_collator = load_data_collator(tokenizer)
```

```python
# Fine tuning and Model training
tokenizer.save_pretrained(output_dir)


model.save_pretrained(output_dir)

training_args = TrainingArguments(
        output_dir=output_dir,
        overwrite_output_dir=overwrite_output_dir,
        per_device_train_batch_size=per_device_train_batch_size,
        num_train_epochs=num_train_epochs,
)

trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=train_dataset,
)

trainer.train()
trainer.save_model()
```

```python
# Summary Prediction
test_query = X_test.iloc[0]
```

```python
input_test_query = test_query
```

```python
text = input_test_query
encoded_ids = tokenizer.encode(text, return_tensors="pt")
```

```python
generated_ids = model.generate(
    encoded_ids, 
    max_length=100,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    top_k=40,
    top_p=0.95, 
    temperature=0.7
)

summary = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(summary)
```
